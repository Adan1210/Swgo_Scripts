```python
import matplotlib.pyplot as plt
import numpy as np
from math import floor
```

```python
xgrid1 = []
ygrid1 = []

p=80/np.sqrt(3) #valor que debe tener a, en sqrt(3)x+4*sqrt(3)*a

d=[]

pcopy=floor(p)

d=np.arange(-pcopy,pcopy+1,1)
```

```python
xgrid2 = []
ygrid2 = []
paridad2 = []
desp_x= 160


z2=np.arange(-desp_x,desp_x+2,2)
```

```python
for i in d:
    for x in z2:
        y=np.sqrt(3)*x+4*np.sqrt(3)*i
        
        if x**2+y**2 < 159**2: #NECESITA EXPLICACION
         xgrid2.append(x)
         ygrid2.append(y)
         paridad2.append((x,y))
```

```python
paridad1 = []

desp_y=296

t=37.5/np.sqrt(3)

tcopy=floor(t)

b=np.arange(-tcopy,tcopy+1,1)

z1=np.arange(-desp_y,desp_y+8,8)
#%%
for i in b:
    for x in z1:
        y=np.sqrt(3)*x+16*np.sqrt(3)*i
        
        if x**2+y**2<296**2 and x**2+y**2>163**2:
            xgrid1.append(x)
            ygrid1.append(y)
            paridad1.append((x,y))
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots()

ax.scatter(xgrid2, ygrid2, s=2, color='red')
ax.scatter(xgrid1, ygrid1, s=3, color='green')
ax.tick_params(axis='both', which='major', labelsize=14)
#ax.axis([-200,200,-200,200])
#plt.savefig('Array SWGO.png',dpi=540)

plt.show()   
```

```python
def trunc(values, decs=0):
    return np.trunc(values*10**decs)/(10**decs)
```
